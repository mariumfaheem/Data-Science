#**LOC** Label Based Selection
#**ILOC** Index Based Selection

```python
data[(data.rating)=='PG']
```

```python
data.loc[(data.rating=='PG-13') & (data["release year"]==2011)]
```

```python
data.loc[(data.rating=='PG-13') & (data["release year"]==2011)]
```

```python
data.loc[(data.rating=="TV-14") &(data["user rating score"]>=90)]
```

#**LOC** : First condition , specific columns 

```python
data.loc[(data.rating=='TV-14'),['title','rating','ratingLevel','ratingDescription','release year']]
```

#**ILOC** : First condition , specific Index

```python
data.iloc[[0,5]]
```

#**ILOC** : first rows indexes , second column indexes

```python
data.iloc[[0,5],0:2]
```

```python
data.iloc[[0,5],:-1]
```

```python
avg_user_rating_score=data["user rating score"].mean(axis=0)
```
